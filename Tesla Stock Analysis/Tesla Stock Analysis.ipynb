# Model Evaluation and Prediction

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

## Simple Linear Regression I

```python
df[["Open","Total Trade Quantity"]].corr()

## Simple Linear Regression I
```

```python
X = df[["Open"]]
Y = df["Total Trade Quantity"]
lr.fit(X,Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
Yhat = lr.predict(X)
lr.predict(X)[0:5]
```

### Mathematical Model

Total Trade Quantity = Coefficient * Open Price + Intercept

Total Trade Quantity = 21077.06220447 * Close Price + 1646673.1449512858

```python
sns.regplot(x="Open", y = "Total Trade Quantity", data = df)
```

```python
sns.residplot(x="Open", y = "Total Trade Quantity", data = df)
```

```python
plt.figure(figsize=(12, 6))

ax1 = sns.distplot(Y, hist=False, color="r", label="Actual Value")
sns.distplot(Yhat, hist=False, color="b", label="Fitted Values" , ax=ax1)

plt.title('Actual vs Fitted Values for Total Trade Quantity')
plt.show()
plt.close()
```

As per the graph comparing predicted values and the actual values of the Total Trade Quantity. It can be interpreted this is a case of overfitting of the model.

## Multiple Linear Regression

```python
M = df[["Open", "Close", "Adj Close", "Low", "High"]]
lr.fit(M,Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
Yhat2 = lr.predict(M)
```

### Mathematical Model

Total Trade Quantity = Coefficient * Open Price + Coefficient * Close Price + Coefficient * Adj Close + Coefficient * Low + Coefficient * High + Intercept

Total Trade Quantity = 27073.645583682224 * Open Price - 10522143991.973726 * Close Price + 10522088229.296082 * Adj Close - 666163.667860508 * Low + 693074.8713555336 * High + 1385823.8330288315

```python
plt.figure(figsize=(12, 6))

ax1 = sns.distplot(Y, hist=False, color="r", label="Actual Value")
sns.distplot(Yhat2, hist=False, color="b", label="Fitted Values" , ax=ax1)

plt.title('Actual vs Fitted Values for Total Trade Quantity')
plt.show()
plt.close()
```

As per the graph comparing predicted values and actual values of the Total Trade Quantity. It clearly shows the difference between the predicted values and actual values in least in all of the three models we have considered. Hence this is the best fit model of for now for our data.
